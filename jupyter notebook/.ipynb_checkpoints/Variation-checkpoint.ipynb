## Graphing the variation experiment round

Finally, let's test our hypothesis that 2048 can be played without using the corner strategy.

Let's begin by plotting the raw data. 

```python
import numpy as np
import pandas as pd
from math import floor
import matplotlib.pyplot as plt

control = pd.read_csv('./data/variation.csv',names=["Best solution"])
control.round(0)
control.astype(int)
control.style.set_caption("Raw control evaluation")

ax = control.plot.line()
ax.set_title("Variation experiment round raw data")
ax.set_xlabel("Generation")
ax.set_ylabel("Best solution")
```

Unlike the rest of the graphs we've examing, there appears to be a slight downward trend towards the end of this graph, but it's difficult to tell looking at the raw data. In the following graph, every generation's value is the average of itself and the 49 preceding values.

```python
import numpy as np
import pandas as pd
from math import floor
import matplotlib.pyplot as plt

control = pd.read_csv('./data/variation.csv',names=["Best solution"])
control.round(0)
control.astype(int)

raw = [floor(row[0]) for row in control.values]
smoothed = []
for (i,s) in enumerate(raw[50:]):
    smoothed.append(sum(raw[i:i+50])/50)

graph = pd.DataFrame({"Best solution": smoothed})
    
ax = graph.plot.line()
ax.set_title("Variation experiment round data smoothed over 50 generations")
ax.set_xlabel("Generation")
ax.set_ylabel("Best solution")
```

We can now see that while scores increase during the first half of training, during the second half they stagnate, and even begin to fall. This is still a very jagged graph, so let's smooth to the average value of the generation and the 249 preceding generations.

```python
import numpy as np
import pandas as pd
from math import floor
import matplotlib.pyplot as plt

control = pd.read_csv('./data/variation.csv',names=["Best solution"])
control.round(0)
control.astype(int)

raw = [floor(row[0]) for row in control.values]
smoothed = []
for (i,s) in enumerate(raw[250:]):
    smoothed.append(sum(raw[i:i+250])/250)

graph = pd.DataFrame({"Best solution": smoothed})

ax = graph.plot.line()
ax.set_title("Variation experiment round data smoothed over 250 generations")
ax.set_xlabel("Generation")
ax.set_ylabel("Best solution")
```

We can now see that past the 6000th generation, little progress is made. There are two possible explanations for this behavior. One is that the model is set up in such a way that past a certain point, it will fail to progress. However, we can rule this explanation out as the **control** round [continued to progress](Control.ipynb) wihout interuption.

The other explanation for this failure is that the machine has neared, or hit, the score ceiling for this strategy.

Important to note that is that if it has indeed reached the skill ceiling for this strategy, it never managed to reach the 2048 tile, as that would award a 1000000 bonus to the fitness function. However, the control experiment round was _also_ unable to play a winning game within 10000 generations.
