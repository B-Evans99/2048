## Graphing the control experiment round

This round used a straight forward fitness function that only considered the game score. It's purpose is to be compared against the **variation experiment round**.

Let's begin by plotting the raw data. 

```python
import numpy as np
import pandas as pd
from math import floor
import matplotlib.pyplot as plt

control = pd.read_csv('./data/evaluate_control.csv',names=["Best solution"])
control.round(0)
control.astype(int)
control.style.set_caption("Raw control round")

ax = control.plot.line()
ax.set_title("Control experiment round raw data")
ax.set_xlabel("Generation")
ax.set_ylabel("Best solution")
```

At first glance, this round is less impressive than it's corresponding evaluation round.

To clarify any trends, in the following graph, every generation's value is the average of itself and the 49 preceding values.

```python
import numpy as np
import pandas as pd
from math import floor
import matplotlib.pyplot as plt

control = pd.read_csv('./data/control.csv',names=["Best solution"])
control.round(0)
control.astype(int)

raw = [floor(row[0]) for row in control.values]
smoothed = []
for (i,s) in enumerate(raw[50:]):
    smoothed.append(sum(raw[i:i+50])/50)

graph = pd.DataFrame({"Best solution": smoothed})
    
ax = graph.plot.line()
ax.set_title("Control experiment round data smoothed over 50 generations")
ax.set_xlabel("Generation")
ax.set_ylabel("Best solution")
```

The trend of increasing scores over time is much clearer now. This is still a very jagged graph, so let's smooth to the average value of the generation and the 249 preceding generations.

```python
import numpy as np
import pandas as pd
from math import floor
import matplotlib.pyplot as plt

control = pd.read_csv('./data/control.csv',names=["Best solution"])
control.round(0)
control.astype(int)

raw = [floor(row[0]) for row in control.values]
smoothed = []
for (i,s) in enumerate(raw[250:]):
    smoothed.append(sum(raw[i:i+250])/250)

graph = pd.DataFrame({"Best solution": smoothed})

ax = graph.plot.line()
ax.set_title("Control experiment round data smoothed over 250 generations")
ax.set_xlabel("Generation")
ax.set_ylabel("Best solution")
```

Interestingly, both the **control** round and the **variation** round have distinct cycles where the generations, which had been increasingly skilled, undergo a period of lesser skill. In this control group, this precedes a rise to a higher best score peak than ever previously seen. However, the outcomes are [different](variation.ipynb) in the **variation** round.
