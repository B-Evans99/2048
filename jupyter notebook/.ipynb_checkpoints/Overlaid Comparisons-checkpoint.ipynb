```python
import numpy as np
import pandas as pd
from math import floor
import matplotlib.pyplot as plt

exp = pd.read_csv('./data/control.csv',names=["Best solution"])[:1000]
exp.round(0)
exp.astype(int)

rawExp = [floor(row[0]) for row in exp.values]
smoothedExp = []
for (i,s) in enumerate(rawExp[50:]):
    smoothedExp.append(sum(rawExp[i:i+50])/50)

eva = pd.read_csv('./data/evaluate_control.csv',names=["Best solution"])[:1000]
eva.round(0)
eva.astype(int)

rawEva = [floor(row[0]) for row in eva.values]
smoothedEva = []
for (i,s) in enumerate(rawEva[50:]):
    smoothedEva.append(sum(rawEva[i:i+50])/50)
    
graph = pd.DataFrame({"Evaluation": smoothedEva, "Experiment": smoothedExp})
ax = graph.plot.line()
ax.set_title("Control rounds")
ax.set_xlabel("Generation")
ax.set_ylabel("Best solution")
```
